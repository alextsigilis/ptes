```python
df = pd.concat([dataN2, dataNotN2], axis=0)
df
```

```python
X = df.iloc[:, 2:9]
y = df.iloc[:,10]
```

```python
y[y != 'Sleep stage N2'] = 0
y[y == 'Sleep stage N2'] = 1
```

```python
plt.hist(y)
```

```python
le = LabelEncoder()
ohe = OneHotEncoder(sparse=False)
X = StandardScaler().fit_transform(X)
y = le.fit_transform(y).reshape((y.shape[0], ))
```

```python
# Apply PCA
pca = PCA(n_components=4)
X = pca.fit_transform(X)
```

```python
plt.plot(pca.explained_variance_, 'o-')
```

```python
# Split training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y)
```

```python
# Instatiate the MLP
mlp = MLPClassifier(hidden_layer_sizes=[16, 16, 32, 32, 64, 32, 32, 16, 16, 8],
                    activation='logistic',
                    solver='adam',
                    learning_rate='constant',
                    max_iter=300,
                    warm_start=True,
                    verbose=True)
```

```python
mlp.fit(X_train, y_train)
```

```python
print(f"Cohen kappa at train dataset = {cohen_kappa_score(y_train, mlp.predict(X_train))}")

y_pred = mlp.predict(X_test)
print(f"Cohen kappa at test dataset = {cohen_kappa_score(y_test, y_pred)}")


C = confusion_matrix(y_test, y_pred, normalize='true')
ConfusionMatrixDisplay(C, display_labels=le.classes_, ).plot()
plt.plot()
```
